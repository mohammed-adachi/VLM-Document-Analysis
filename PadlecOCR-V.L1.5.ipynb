```python
%%bash
source /content/py310/bin/activate

# J'ai enlevé le "-c" ici. On appelle juste "python" et on lui donne le bloc EOF.
python <<EOF
from paddleocr import PaddleOCRVL

try:
    pipeline = PaddleOCRVL()
    output = pipeline.predict("https://paddle-model-ecology.bj.bcebos.com/paddlex/imgs/demo_image/paddleocr_vl_demo.png")
    for res in output:
        res.print()
        res.save_to_json(save_path="output")
        res.save_to_markdown(save_path="output")
except ImportError:
    print("Erreur: PaddleOCRVL n'est pas trouvé. Vérifiez l'installation.")
except Exception as e:
    print(f"Une erreur est survenue : {e}")
EOF
```

```python
%%bash
# 1. Installation OBLIGATOIRE de la version "all" pour avoir PaddleOCRVL
# Sans le crochet [all], Python ne trouvera pas le module PaddleOCRVL
source /content/py310/bin/activate
# 2. Exécution de votre script avec PaddleOCRVL
python <<EOF
import os
from paddleocr import PaddleOCRVL

# Nom de votre image (assurez-vous qu'elle est bien dans le dossier Colab)
img_path = "image.png"

if not os.path.exists(img_path):
    print("ERREUR: L'image 'test_image.jpg' est introuvable !")
    print("Veuillez glisser votre image dans le dossier à gauche et la renommer.")
else:
    print("Chargement du modèle PaddleOCR-VL (cela peut prendre une minute)...")
    # Initialisation du modèle Vision-Language
    pipeline = PaddleOCRVL()

    print(f"Analyse en cours de {img_path}...")
    output = pipeline.predict(img_path)

    # Sauvegarde des résultats comme vous l'avez demandé
    for res in output:
        res.print()  # Affiche le résultat dans la console
        res.save_to_json(save_path="output")
        res.save_to_markdown(save_path="output")

    print("Succès ! Les fichiers JSON et Markdown sont dans le dossier 'output'.")
EOF
```
